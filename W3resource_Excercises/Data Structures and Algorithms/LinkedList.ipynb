```python
# Node Creation Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
# Linked List Class
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertAtBegin(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def insertAtEnd(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        
        temp_node = self.head
        while temp_node.next != None:
            temp_node = temp_node.next
    
        temp_node.next = new_node
        
    def insertAtIndex(self, data, index):
        
        if index == 0:
            self.insertAtBegin(data)
            return
        
        temp_node = self.head
        pos = 0
        
        while pos+1 != index and temp_node != None:
            pos += 1
            temp_node = temp_node.next
        
        if temp_node is None:
            print("index not present")
        else:
            new_node = Node(data)
            new_node.next = temp_node.next
            temp_node.next = new_node
            
    def update_node(self, data, index):
        
        temp_node = self.head
        pos = 0
        
        while pos != index and temp_node != None:
            pos += 1
            temp_node = temp_node.next
        
        if temp_node is None:
            print("index not present")
        else:
            temp_node.data = data
            
    def remove_first_node(self):
        
        if self.head == None:
            return
        
        self.head = self.head.next
        
    def remove_last_node(self):
        
        if self.head is None:
            return
        
        if self.head.next is None:
            self.head = None
            return
        
        temp_node = self.head
        
        while temp_node.next.next is None:
            temp_node = temp_node.next
            
        temp_node.next = None
        
    def remove_index_node(self, index):
        
        if self.head is None:
            return
        
        if index == 0:
            self.remove_first_node()
            return
            
        temp_node = self.head
        pos = 0
        
        while temp_node != None and pos+1 != index:
            temp_node = temp_node.next
            pos += 1
                        
        if temp_node is None or temp_node.next is None:
            print("Index doesn't exists")
            return
        
        temp_node.next = temp_node.next.next
        
    def printLL(self):
        temp_node = self.head
        
        while temp_node:
            print(temp_node.data)
            temp_node = temp_node.next
    
        
```

```python
# create a new linked list
llist = LinkedList()

# add nodes to the linked list
llist.insertAtEnd('a')
llist.insertAtEnd('b')
llist.insertAtBegin('c')
llist.insertAtEnd('d')
llist.insertAtIndex('g', 2)

# print the linked list
print("Node Data:")
llist.printLL()

# remove nodes from the linked list
print("\nRemove First Node:")
llist.remove_first_node()
llist.printLL()

print("\nRemove Last Node:")
llist.remove_last_node()
llist.printLL()

print("\nRemove Node at Index 1:")
llist.remove_index_node(1)
llist.printLL()

# print the linked list after all removals
print("\nLinked list after removing a node:")
llist.printLL()

print("\nUpdate node Value at Index 0:")
llist.update_node('z', 0)
llist.printLL()
```
