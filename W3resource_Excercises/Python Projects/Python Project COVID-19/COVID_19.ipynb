```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
covid_data= pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-19-2020.csv')

covid_data
```

```python
covid_data['Last Update'] = pd.to_datetime(covid_data['Last Update'])

# result_data = covid_data.groupby('Last Update')
date_index = covid_data.set_index('Last Update')
date_index.sort_index(inplace=True)
date_index.reset_index(inplace=True)
date_index

sorted_date_group = date_index.groupby('Last Update').sum()
sorted_date_group.reset_index(inplace=True)
# sorted_date_group['Last Update'][0]
sorted_date_group.plot(x='Last Update', y='Confirmed', title='Confirmed COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()  # To make sure labels do not overlap
plt.show()
```
