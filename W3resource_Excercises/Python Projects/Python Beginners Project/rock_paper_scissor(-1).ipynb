```python
# importing modules
import random

from collections import Counter
```

```python
# signs and their values
values = {
    "1" : "Rock",
    "2" : "Paper",
    "3" : "Scissors"
}
```

```python
class Player:
    def __init__(self, name):
        self.player_name = name
        
    def make_choice(self, one_hand = random.randint(1,3), second_hand =random.randint(1,3)):
        if type(one_hand) == type(1):
            return (str(one_hand), str(second_hand))
        
        return (one_hand, second_hand)
```

```python
class RPS:
    def __init__(self, rounds):
        self.rounds = rounds
       
    def show_choices(self, player1_choices, player2_choices):
        print("\nPlayer1 Choices are:")
        for choice in player1_choices:
             print(values[choice], end=" ")
             
        print("\nPlayer 2 Choices are:")
        for choice in player2_choices:
            print(values[choice], end=" ")
            
    def remove_choices(self, player1_choices, player2_choices, player1_choice):
        
        player1_list = []
        player2_list = []
        
        if player1_choices.count(player1_choice) == 2:
            player1_list.append(values[player1_choice])
        else:
            for choice in player1_choices:
                if choice != player1_choice:
                    player1_list.append(values[choice])
                
        for choice in player2_choices:
            player2_list.append(values[choice])
            
        player2_list.pop()        
        return "".join(player1_list), "".join(player2_list)
    
    def evaluate_choices(self, player1, player2):
        
        print(f"\nPlayer1 Choice was: {player1}")
        print(f"Player2 Choice was: {player2}")
        
        if (player1 == "Rock" and player2 == "Scissors") or (player1 == "Paper" and player2 =="Rock") or (player1 == "Scissors" and player2 == "Paper"):
            return "Player 1 Wins"
        elif (player2 == "Rock" and player1 == "Scissors") or (player2 == "Paper" and player1 =="Rock") or (player2 == "Scissors" and player1 == "Paper"):
            return "Player 2 Wins"
        else:
            return "Draw"     
             
        
        
```

```python
rounds = int(input("Enter the rounds you wanna play??"))

rps = RPS(rounds)

player1_name = input("Etner player 1 name:")
player2_name = "Bot"

player1 = Player(player1_name)
player2 = Player(player2_name)

final_result = []
while rps.rounds > 0:

    player1_one_hand = input("Enter Rock/Paper/Scissor from 1/2/3/ for first hand: ")
    player1_second_hand = input("Enter Rock/Paper/Scissor from 1/2/3/ for second hand: ")

    player1_choices = player1.make_choice(player1_one_hand, player1_second_hand)
    player2_choices = player2.make_choice()
    
    rps.show_choices(player1_choices, player2_choices)
    
    player1_remove_choice = input("Enter the choice you want to remove: ")
    
    player1_choice, player2_choice = rps.remove_choices(player1_choices, player2_choices, player1_remove_choice)
    
    # print("Final Choices:")
    # print(f"Player 1: {player1_choice}, Player 2: {player2_choice}")
    result = rps.evaluate_choices(player1_choice, player2_choice)
    print(result)
    if result == "Player 1 Wins":
       final_result.append(1)
    elif result == "Player 2 Wins":
        final_result.append(2)
    else:
        final_result.append(0)
        
    rps.rounds -= 1
  
win_list =  Counter(final_result)
win_list
for winner, wins in win_list.items():
    if wins > int(rounds/2):
        if winner == 0:
            print("\n\n\nNeither Wins, it's a Draw!!!!!!")
        else:
            print(f"\n\n\nPlayer {winner} wins the Series!!!!!!!!!!!!!!!!")
    


    
    
```
