```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/movies_metadata.csv', encoding='ISO-8859-1', on_bad_lines='skip')
df.head()
```

```python
# Write a Pandas program to display the movies (title, number of votes) that received specified number of votes.
result = df.loc[df.vote_count > 2000]

result[['original_title','vote_count']] 
```

```python
# Write a Pandas program to display the movies (title, runtime) longer than 30 minutes and shorter than 360 minutes.

result = df.loc[(df.runtime > 30) & (df.runtime < 360)]

result[['original_title','runtime']] 
```

```python
# Write a Pandas program to sort movies on runtime in descending order.
new_df = df.set_index('runtime')
new_df.sort_index(ascending=False)
```
