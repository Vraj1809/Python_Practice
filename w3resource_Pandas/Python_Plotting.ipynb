```python
import pandas  as pd
import matplotlib.pyplot as plt
```

```python
# Write a  Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

df = pd.read_csv('alphabet_stock_data.csv')

df.Date = pd.to_datetime(df.Date)
df

plt.figure(figsize=(5,5))
plt.suptitle('Stock prices of Alphabet Inc.,\n01-04-2020 to 30-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ price", fontsize=16, color='black')
 
df['Close'].plot(color='green')
plt.show()
```

```python
# Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.

df = pd.read_csv('alphabet_stock_data.csv')
df

# between Specific Dates
# start date 1 May 
# end Date 18 Sep
start_date = pd.to_datetime('1/5/2020', format="%d/%m/%Y", errors="coerce")
end_date = pd.to_datetime('18/9/2020', format="%d/%m/%Y", errors="coerce")

df.Date = pd.to_datetime(df.Date)

new_def = df[(df.Date >= start_date) & (df.Date <= end_date)]
new_def

plt.figure(figsize=(5,5))
new_def.plot(x='Date', y=['Open','Close'])
plt.suptitle('Opening and Closing Stock prices of Alphabet Inc.,\n01-05-2020 to 18-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("Price", fontsize=16, color='black')
plt.show()
```
