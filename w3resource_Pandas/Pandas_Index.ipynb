```python
import pandas as pd
```

```python
# Write a Pandas program to display the default index and set a column as an Index in a given dataframe.

df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4'],
    't_id':['t1', 't2', 't3', 't4', 't5', 't6']})

print("Data Frame Display using the Default Index: \n", df)

# df1 = df.set_index('school_code')
# print("\nSetting School code the default index: \n", df1)
# print("\nnot inplace so the df is not changed: \n", df)

df.set_index('school_code', inplace=True)
print("\nUsing Inplace so the main df is changed: \n", df)
print("Accessing using the new index: \n", df.loc['s001'])

df.reset_index(inplace=True)
print("\nIndex is reset: \n", df)
```

```python
# Write a Pandas program to create a multi Index frame using two columns and using an Index and a column.

df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4'],
    't_id':['t1', 't2', 't3', 't4', 't5', 't6']})

print("Data Frame Display using the Default Index: \n", df)

df.set_index(['school_code','t_id'], inplace=True)
print("\nUsing Inplace so the main df is changed: \n", df)
print("Accessing using the new index: \n", df.loc['s001'])

df.reset_index(inplace=True)
print("\nIndex is reset: \n", df)
```

```python
# Setting the Columns as the index and then taking them back to the dataframe and resting the index
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_of_birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32],
    't_id' : ['t1', 't2', 't3', 't4', 't5', 't6']})

# print("Original DataFrame: \n", df)
# print("Original Indexes: ", df.index)

df.set_index(['t_id', 'school_code', 'class'], inplace=True)
# print("With Changed index: \n", df)
# print("Indexes: \n", df.index)

# the levels of index are set as level 0 -> t_id, level 1 -> school_code, level 2 -> class
df.reset_index(level = 0, inplace=True)
print("With Reset the level 0 index: \n", df)

df.reset_index(level='class', inplace=True)
print("With reseting the index using the column name: \n", df)

df.reset_index()
```
